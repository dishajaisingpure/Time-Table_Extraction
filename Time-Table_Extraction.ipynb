```python
# Importing all the necessary packages

import pandas as pd
import glob
```

```python
""" Reading all the files from the specified location, searching for the name of the teacher and if the name 
is found adding it to a new dataframe. """

output_table = pd.DataFrame() 
for files in glob.iglob('.\Timetable input sheets\*.xlsx'):
    xls = pd.ExcelFile(files)
    sheet_to_df_map = {}
    for sheet_name in xls.sheet_names:
        sheet_to_df_map[sheet_name] = xls.parse(sheet_name)
    for i in sheet_to_df_map:
        grade_div = []
        data = pd.DataFrame(sheet_to_df_map[i].values)
        print(data)
        for i in data.loc[2]:
            if pd.isna(i) == False:
                grade_div.append(i)
        print(grade_div)
        df = pd.DataFrame(data.loc[4:])
        df = df.rename(columns = {0: 'Day / Date', 1: 'Time', 2: 'Teacher', 3: 'Subject'}, inplace = False)
        df = df[['Day / Date','Time','Teacher','Subject']]
        df = df.dropna(how='all') 
        df['Teacher'] = df['Teacher'].str.replace(' ','')
        df['Teacher'] = df['Teacher'].fillna('None')
        df = df.reset_index(drop=True)
        val = ""
        for n,i in enumerate(df['Day / Date']):
            if n % 2 == 0:
                val = i
            else:
                df['Day / Date'][n] = val
        print(df)
        df1 = df[df["Teacher"].str.contains('Ms.Archana')]
        if df1.empty == False:
            df1 = df1.drop('Teacher',1)
            df1.insert(2, "Grade", grade_div[0].partition(":")[2]) 
            df1.insert(3, "Division", grade_div[1].partition(":")[2]) 
            print(df1)
    #         print('yes')
            output_table = output_table.append(df1, ignore_index = True) 
```

```python
"""After all the files have been iterated and the dataframe entries are complete, 
the dataframe is sorted; first according to the Date and then acoording to the time. 
Then the dataframe is written to the output excel file."""

output_table = output_table.sort_values(['Day / Date','Time'])
print(output_table)
output_table.to_excel(".\\Timetable input sheets\\timetable output.xlsx", index=False) 
```
